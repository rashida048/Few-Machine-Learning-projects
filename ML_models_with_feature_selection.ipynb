```python
def root_mean_squared_error(pred, target):
    return np.sqrt(((pred - target)**2).mean())
```

```python
def rae(actual: np.ndarray, predicted: np.ndarray):
    """ Relative Absolute Error (aka Approximation Error) """
    return np.sum(np.abs(actual - predicted)) / (np.sum(np.abs(actual - np.mean(actual))) + 0.01)
rae(y_test, y_pred)
```

```python
def rrse(actual: np.ndarray, predicted: np.ndarray):
    """ Root Relative Squared Error """
    return np.sqrt(np.sum(np.square(actual - predicted)) / np.sum(np.square(actual - np.mean(actual))))

rrse(y_test, y_pred)
```

```python
def eval_matrix(clf, X, y):
    y_pred = clf.predict(X)
    acc = clf.score(X, y)
    cps = cohen_kappa_score(y, y_pred)
    mean_abs_err = mean_absolute_error(y_pred, y)
    rmse = root_mean_squared_error(y_pred, y)
    rel_abs_err = rae(y, y_pred)
    root_rse = rrse(y, y_pred)
    return {'accuracy': acc, 'inaccuracy': 1-acc,
           'Kappa_stat': cps,
           'mean_absolute_erro': mean_abs_err,
           'root_mean_squared_erro': rmse,
           'relative_absolute_error': rel_abs_err,
           'root_relative_squared_error': root_rse,
           'total_number_of_instances': len(y)}
```

```python
def confusion_matrix_scorer(clf, X, y):
    y_pred = clf.predict(X)
    cm = confusion_matrix(y, y_pred)
    return {'tn': cm[0, 0], 'fp': cm[0, 1],
            'fn': cm[1, 0], 'tp': cm[1, 1]}
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve
```

```python
def matrix(clf, X, y):
    y_pred = clf.predict(X)
    d = confusion_matrix_scorer(clf, X, y)
    tn, fp, fn, tp = d['tn'], d['fp'], d['fn'], d['tp']
    tpr2 = tn/(tn+fn)
    fpr2 = fn/(fn+tp)   
    precision2 = tn/(tn+fn)
    recall2 = tn/(tn+fp)
    f_2 = 2*precision2*recall2/(precision2 + recall2)
    mcc2 = (tn*tp - fn*fp)/np.sqrt((tn+fn)*(tn+fp)*(tp+fn)*(tp+fp))
    #roc_area1 = roc_auc_score(y, y_pred)
    fpr22, tpr22, thresholds22 = metrics.roc_curve(y, y_pred, pos_label=2)
    roc_area2 = metrics.auc(fpr22, tpr22)
    #prc_area1 = average_precision_score(y, y_pred)
    precision22, recall22, thresholds2 = precision_recall_curve(y_test, y_pred, pos_label = 2)
    pr_auc2 = metrics.auc(recall22, precision22)
        
    tpr1 = tp/(tp+fp)
    fpr1 = fp/(fp+tn)
    precision1 = tp/(tp+fp)
    recall1 = tp/(tp+fn)
    f_1 = 2*precision1*recall1/(precision1 + recall1)
    mcc1 = (tn*tp - fn*fp)/np.sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))
    #roc_area1 = roc_auc_score(y, y_pred)
    fpr11, tpr11, thresholds11 = metrics.roc_curve(y, y_pred, pos_label=1)
    roc_area1 = metrics.auc(fpr11, tpr11)
    #prc_area2 = average_precision_score(y, y_pred)
    precision11, recall11, thresholds1 = precision_recall_curve(y_test, y_pred, pos_label=1)
    pr_auc1 = metrics.auc(recall11, precision11)
    
    dict2 = {'tpr1':tpr2, 'fpr1': fpr2, 'precision1': precision2, 'recall1': recall2, 
             'f_measure1': f_2, 'mcc1': mcc2, 'roc_area1': roc_area2, 'pr_area1': pr_auc2}
    dict1 = {'tpr2':tpr1, 'fpr2': fpr1, 'precision2': precision1, 'recall2': recall1, 
             'f_measure2':f_1, 'mcc2': mcc1, 
            'roc_area2': roc_area1, 'pr_area2': pr_auc1}
    
    #dict3 = {"roc_area": roc_auc_score(y, y_pred), "prc_area": average_precision_score(y, y_pred)}
    return (d, dict1, dict2)
```

### Wrapper Method

```python
import statsmodels.api as sm
X_1 = sm.add_constant(X)
model = sm.OLS(y, X_1).fit()
model.pvalues
```

```python
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```
